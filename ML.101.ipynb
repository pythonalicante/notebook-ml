## Obtención de datos
---
Utilizaremos un conjunto de datos que nos proporciona la propia librería `scikit-learn`:

```python
from sklearn.datasets import load_boston

datos_casas = load_boston()
# print(cosa.DESCR)  # Descomentar para obtener información más detallada sobre el conjunto de datos
```

Para encapsular los datos utilizaremos la librería `Pandas`:

Por simplificar el ejemplo, nos quedaremos tan solo con 5 características. Para ello, primero se declarará una tabla con
todas las características y luego, descartaremos aquellas que no entraran dentro de este ejemplo.

```python
import pandas as pd


caracteristicas = {'CRIM':0, 'ZN':1, 'INDUS':2, 'CHAS':3, 'NOX':4, 'RM':5, 'AGE':6, 'DIS':7, 'RAD':8, 'TAX':9, 'PTRATIO':10, 'B':11, 'LSTAT':13}
caracteristicas_casas = pd.DataFrame(datos_casas.data, columns=caracteristicas)


caracteristicas_casas
```

Por otro lado, definimos la característica **objetivo** que queremos predecir, es decir, el precio.

```python
precios_casas = pd.Series(datos_casas.target)
precios_casas
```

## Primer modelo: regresión lineal
---
Para declarar este modelo de predicción, utilizando `scikit-learn`, primero instanciamos el modelo y, a continuación, le pasamos los datos de los que tiene que aprender:

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression(normalize=True)
lr.fit(caracteristicas_casas, precios_casas)
```

## Analizando los resultados
---
Para observar la predicción del modelo lineal, así como, el efecto que tiene cada característica al precio. Para ello, la mejor forma es visualizando los resultados utilizando `matplotlib`.

```python
%matplotlib inline
import matplotlib.pyplot as plt

def plot_feature(caracteristica):
    # Valores reales
    plt.scatter(caracteristicas_casas[caracteristica], precios_casas, c='b', alpha=0.3)
    # Valores predecidos
    plt.plot(caracteristicas_casas[caracteristica],
             caracteristicas_casas[caracteristica] * lr.coef_[caracteristicas[caracteristica]] + lr.intercept_, 'k')
    # Etiquetas
    plt.legend(['Predicted value', 'Actual value'])
    plt.xlabel(caracteristica)
    plt.ylabel("Median value in $1000's")
```

Según índice de criminalidad

```python
plot_feature('CRIM')
```

Según el nivel de polución

```python
plot_feature('NOX')
```

Según el número de habitaciones

```python
plot_feature('RM')
```

Según la antigüedad

```python
plot_feature('AGE')
```
